```python
# importer les packages
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv("/content/drive/MyDrive/Simple_regression_TP/Multiple_Linear_Regression_TP/Fuel_Consumption.csv")
```

```python
df.columns
```

```python
df1=X =df.drop(["CYLINDERS","FUELCONSUMPTION_COMB"],axis=1)
df1.shape
df1
```

```python
# Fractionner les données
X1 =df1.ENGINESIZE.values
y1 = df1["CO2EMISSIONS"].values
```

```python
# Import la fonction train_test_split
from sklearn.model_selection import train_test_split
X1_train,X1_test,y1_train,y1_test=train_test_split(X1,y1,test_size=0.2,random_state=4)
```

```python
# Transformer X_train et X_test en 2D matrice 
X1_train =  X1_train[:,np.newaxis] 
X1_test = X1_test[:,np.newaxis]
```

```python
# Importer linear regression
from sklearn.linear_model import LinearRegression
# Instantier le modèle 
lr = LinearRegression()
# Entrainer le modèle
lr.fit(X1_train,y1_train)
```

```python
# Les coefficients
print('Coefficients: ', lr.coef_)
print('Intercept: ', lr.intercept_)
```

```python
# Prédire le  0 des données test en utilisant le coefficient (pente) et l'intercept(l'ordonnée d'origine)
y_pred_model1 = lr.coef_*X1_test[:1]+lr.intercept_
```

```python
# predire pour les donnees test
y_pred_model1=lr.predict(X1_test)
y_pred_model1
```

```python
#
from sklearn.metrics import r2_score,mean_squared_error
# Calculer le R2
rsq = r2_score(y_pred_model1,y1_test)
rsq
```

```python
# calculer le Root Mean square error
from math import sqrt
rmse = sqrt(mean_squared_error(y_pred_model1,y1_test))
```

```python
rmse
```

```python
import matplotlib.pyplot as plt
plt.scatter(df.ENGINESIZE,df.CO2EMISSIONS, color="r")
plt.plot()
plt.scatter(df.ENGINESIZE, df.CO2EMISSIONS,  color='r')
plt.plot(X1_train, lr.coef_[0]*X1_train + lr.intercept_, color = 'b')

```
